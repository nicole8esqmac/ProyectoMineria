## Total de sintomas 

```python
InicioSintoma = pd.read_csv("C:/Users/Nicole/Desktop/ProyectoFinalMineriaCOVID19/Inicio de síntomas.csv")
TotalInicioSintoma=InicioSintoma
TotalInicioSintoma.drop(['departamento','codigo_departamento','codigo_municipio'], axis = 'columns', inplace=True)
# np.sum(TotalInicioSintoma,axis=1)#suma de reglones en una matriz

TotalInicioSintoma.loc[:,'Total Sintoma']=TotalInicioSintoma.sum(axis=1)
TotalInicioSintoma.head()
```

* Se puede observar los datos impresos de las columnas municipio y las fecha de inicio de síntomas del 2020-02-13 al 2021-05-26
* Asi mismo se agregó una nueva columna "Total Sintoma"  se hizo sumatoria de los reglones de todas fechas es decir la totalidad  de los sintomas de cada ciudadano por municipio 

```python
# Extraer más de una columna para un subconjunto
sintoma = TotalInicioSintoma[["municipio","poblacion", "Total Sintoma"]]
sintoma.head(20)
```

## Gráfica de Sintomas por municipio 

```python
# x = list('{0:.0%}'.format(sintoma["municipio"]), ha="center")

X = list(sintoma["municipio"])
Y = list(sintoma["Total Sintoma"])


fig, axs = plt.subplots(figsize=(50,10))
axs.bar(X,Y)

plt.xlabel("Municipio")
plt.ylabel("Ciudadano con Sintomas")
plt.title("SINTOMAS POR MUNICIPIO")
```

* Se puede visualizar en la gráfica de barras que los sintomas de COVID-19 el municipio de Sololá, Panajachel, Nahuala, Santa Lucia Utalan,Santiago Atitlán y San Lucas Toliman son los lugares con un rango de  114 a 744.
* Y el resto de municipios tienen bajo rango no pasa de los 100 ciudadanos con inico de sintomas

```python
sintoma2 = sintoma[sintoma["Total Sintoma"]>=100]
sintoma2.head(20)
```

```python
plt.pie(sintoma2['Total Sintoma'], 
       labels=sintoma2['municipio'],autopct="%1.2f%%")
plt.title("SINTOMAS POR MUNICIPIO MAYOR O IGUAL A  100")
# plt.legend()
plt.show()
```

* Se observa que el porcentaje del municipio de Sololá tiene un mayor dato que las demas  de un 42.98% que en números da un dato de 744, esta cantidad rebasa exageradamneto el 100 de personas con inicio de sintomas , entonces demuestra que el municipio de Sololá tiene un indice mayor de casos con sintomas  le sigue el municipio de Panajachel con un 24.44% que equivale unos 423
